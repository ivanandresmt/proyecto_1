### 4.1 Declarar `ColumnTransformer`

```python
#Antes de colar los transformer vamos a poner las variables mas relevantes en un dataset 

df_modelo = df_union[['fraud_bool','current_address_months_count','credit_risk_score','name_email_similarity','employment_status','source','segmentacion_etaria']]

print(df_modelo.head())
print(df_modelo.describe())

#podemos ver que los valores no son de la misma escala, por lo que vamos a estandarizar los valores para poder tener mejores resultados.

#Para los valores categoricos vamos a usar One-Hot, dado que los valores no son de tipo ordinal 

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


one_hot_encoder = Pipeline(
  steps=[
    ('one_hot', OneHotEncoder(handle_unknown='ignore'))
  ]
)


scaler = Pipeline(
  steps=[
    ('standardize', StandardScaler())
  ]
)
```

### 4.2 Transformar datos

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
  transformers=[
    ('categorical', one_hot_encoder, ['employment_status','source','segmentacion_etaria']),
    ('num', scaler, ['current_address_months_count','credit_risk_score','name_email_similarity'])
  ]
)
```

```python
#Vamos a usar un random forest para este proyecto
from sklearn.ensemble import RandomForestClassifier

rf = Pipeline(
  steps=[
    ('preprocessor', preprocessor),
    ('classifier',  RandomForestClassifier())
  ]
)
```

```python
#Separo el dataset
from sklearn.model_selection import train_test_split

X = df_modelo.drop('fraud_bool', axis=1)
y = df_modelo['fraud_bool']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score

y_pred = rf.predict(X_test)
print("Score : {:2.1f}%".format(f1_score(y_test, y_pred) * 100))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

#el resultado obtenido es normal dado que hay un sobre-entrenamiento de los datos, para corregir este ambito habria que balanciar las clases.
```

```python
#Ahora vamos a calcular el Ratio
print(labels)

# Calcular el número de ejemplos predichos correctamente como fraude
correct_pred_fraud = ((labels == -1) & (y == 1)).sum()

# Calcular el número total de datos predichos como fraude
total_pred_fraud = (labels == -1).sum()

# Calcular el número total de ejemplos que eran realmente fraude
total_real_fraud = (y == 1).sum()

# Calcular el ratio P
ratio_p = correct_pred_fraud / total_pred_fraud

# Calcular el ratio R
ratio_r = correct_pred_fraud / total_real_fraud

print("Ratio P:", ratio_p)
print("Ratio R:", ratio_r)
```
